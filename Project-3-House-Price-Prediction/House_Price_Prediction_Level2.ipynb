# Predicting House Prices with Linear Regression
Internship Project - Data Analytics



**Step 1: Data  Collection**

```python
import pandas as pd
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.shape
```

**Step 2: Data Exploration & Cleaning**


```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

**Step 3: Feature Selection**

```python
y = df['price']
```

```python
x = df[['area','bedrooms','bathrooms','stories','parking']]
x.head()
```

**Step 4: Model Training**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

**Step 5: Model Evaluation**

```python
y_pred = model.predict(x_test)
```

```python
from  sklearn.metrics import mean_squared_error,r2_score

mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

mse,r2
```

**Step 6: Visualization**

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

# Conclusion
This project demonstrates the use of Linear Regression to predict house prices based on property features.
The model shows a reasonable relationship between actual and predicted prices, indicating that features like area, number of bedrooms, and bathrooms significantly influence house prices.
